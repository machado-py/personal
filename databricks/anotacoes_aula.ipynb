```python
%py
# LER AQUIVO QUE ESTÁ NO S3 -----------------------------------------------------------------------------------

endereco = "dbfs:/mnt/???-??????-letscode/raw/top1000_grossing_movies.csv"
df_movies = spark.read.option("header", "true").csv(endereco)
df_movies.display()
```

```python
%py
# CRIAR 3 PASTAS NO S3 -----------------------------------------------------------------------------------
dbutils.fs.mkdirs("/mnt/???-??????-letscode/curated/top_5_movies/")
dbutils.fs.mkdirs("/mnt/???-??????-letscode/curated/gross_genre/")
dbutils.fs.mkdirs("/mnt/???-??????-letscode/curated/least_5_liked_movies/")
```

```python
%py
df_movies = df_movies.withColumnRenamed("Movie Title", "Title")
df_movies = df_movies.withColumnRenamed("Year of Realease", "Year")
df_movies = df_movies.withColumnRenamed("Movie Rating", "Rating")
```

```python
%py
from pyspark.sql.functions import regexp_replace

df_movies = df_movies.withColumn('Gross', regexp_replace('Gross', '\$', ''))
df_movies = df_movies.withColumn('Gross', regexp_replace('Gross', 'M', ''))
df_movies = df_movies.withColumn('Year', regexp_replace('Year', 'X', ''))
df_movies = df_movies.withColumn('Gross', df_movies['Gross'].cast("float").alias('Gross'))
df_movies.createOrReplaceTempView("movies")

df_movies.display()
```

```python
%sql
-- top_5_movies: Salvar os 5 maiores filmes de bilheteria (maior renda), por ano

select Title, Year, Gross from 
    (select *, row_number() OVER (PARTITION BY Year ORDER BY Gross DESC) as rn
     FROM movies) tmp 
     where rn <= 5
     order by Year Desc
```

```python
%py
top5_result = spark.sql("""
    select Title, Year, Gross from 
    (select *, row_number() OVER (PARTITION BY Year ORDER BY Gross DESC) as rn
     FROM movies) tmp 
     where rn <= 5
     order by Year Desc""")
```

```python
%py
# salvando resultado no s3
top5_result.write.format('delta').mode('append').save("/mnt/???-??????-letscode/curated/top_5_movies/")
```

```python
%sql
-- gross_genre: A soma do valor por genero

select Genre, sum(Gross) as TotalGross from movies
group by Genre
order by 2 Desc
```

```python
%py
bygenre_result = spark.sql("""
    select Genre, sum(Gross) as TotalGross from movies
    group by Genre
    order by 2 Desc
""")

bygenre_result.write.format('delta').mode('append').save("/mnt/???-??????-letscode/curated/gross_genre/")
```

```python
%sql
-- least_5_liked_movies: Os 5 piores filmes entre todos de acordo com a critica

select Title, Rating  from movies
order by 2 Asc
limit 5
```

```python
%py
leastlike_result = spark.sql("""
    select Title, Rating  from movies
    order by 2 Asc
    limit 5
""")

leastlike_result.write.format('delta').mode('append').save("/mnt/???-??????-letscode/curated/least_5_liked_movies/")
```

```python
url = 'jdbc:postgresql://??-???.?????????.sa-east-1.rds.amazonaws.com/movies'
user = 'postgres'
senha = '?????'
driver = 'org.postgresql.Driver'
```

```python
def run_query(query, limit=True):
    if limit:
        query = f'{query} limit 5'
    try:
        df = spark.read.format("jdbc")\
            .option("driver", driver)\
            .option("url", url)\
            .option("user", user)\
            .option("password", senha)\
            .option("query", query)\
            .load()
    except Exception as e:
        print(f'Erro na função ExecutarConsulta: {e}')

    return df
```

```python
df = run_query("select * from film")
df.display()
```

```python
df.write.format('delta').mode('append').save("/raw/dvdrental/film/")
```
